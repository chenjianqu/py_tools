```python
%matplotlib inline
from nuscenes.nuscenes import NuScenes

nusc = NuScenes(version='v1.0-mini', dataroot='/media/cjq/新加卷2/BaiduNetdiskDownload/mini/v1.0-mini', verbose=True)
```

```python
nusc.list_scenes()
list_scenes = nusc.scene
print(type(list_scenes))
print(len(list_scenes))

print(type(list_scenes[0]))
print(list_scenes[0])
```

```python
# 获取所有的样本数量
print(f"num_samples:{len(nusc.sample)}")

# 获取第一个样本包含的数据和标注
nusc.list_sample(nusc.sample[0]["token"])

# 可视化这个样本的lidar点云，默认投影到前相机
nusc.render_pointcloud_in_image(nusc.sample[0]['token'], pointsensor_channel='LIDAR_TOP')
```

```python
# 获取数据集中所有的数据数量，包括所有的模态
print(f"sample_data:{len(nusc.sample_data)}")

# 输出第一个样本的信息
print(nusc.sample_data[0])
    
```

```python
scene0=list_scenes[0]

# 获得该scene的起始sample和终止sample
print(scene0["first_sample_token"])
print(scene0["last_sample_token"])
# 获得该scene的sample的数量
print(scene0["nbr_samples"])

# 获得该scene第一个样本的信息
first_sample_token=scene0["first_sample_token"]
my_sample = nusc.get('sample', first_sample_token)
print(my_sample.keys())

print(my_sample["token"])
print(my_sample["timestamp"])
print(my_sample["prev"])
print(my_sample["next"])
print(my_sample["scene_token"])

my_sample_data = my_sample["data"] # dict{name:token}
my_sample_anns = my_sample["anns"] # list[token]

print(my_sample_data.keys())
print(len(my_sample_anns))

print(my_sample_data["CAM_FRONT"])
```

```python
# 获取第一个sample和第一个sample之间的时间戳
first_sample_token=scene0["first_sample_token"]
sample0 = nusc.get('sample', first_sample_token)
sample1 = nusc.get('sample', sample0["next"])

print(sample0["timestamp"])
print(sample1["timestamp"])

sample0_timestamp = sample0["timestamp"]/1000000
sample1_timestamp = sample1["timestamp"]/1000000
print(sample0_timestamp)
print(sample1_timestamp)
```

```python
# 获取该sample前相机的图像
my_sample_cam_front = nusc.get('sample_data', my_sample_data["CAM_FRONT"])
print(my_sample_cam_front)
```

```python
# 获取某个时刻的车辆位姿
ego_pose = nusc.get('ego_pose', my_sample_cam_front["ego_pose_token"])
print(ego_pose)

# 获取相机的内参和外参
calibrated_sensor = nusc.get('calibrated_sensor', my_sample_cam_front["calibrated_sensor_token"])
print(calibrated_sensor)
#外参
print(calibrated_sensor["translation"])
print(calibrated_sensor["rotation"])

# 获取内参
print(calibrated_sensor["camera_intrinsic"])
```
